```python
# year treads for number of occurrences on different charges
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv('./court_prep_2.csv')

sns.displot(data=df, x='charge', hue='Year', multiple='dodge', shrink=.8, height=5, aspect=2)
```

```python
# court_d statistics and court_d trends on different charges each year
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# court_d statistics
grouped_single = df.groupby('charge').agg({'court_d': ['mean', 'min', 'max']})
print(grouped_single)

# court_d trends on different charges each year
court_d_sum = df.groupby(['Year', 'charge'], as_index=False).sum()
court_d_sum = court_d_sum.drop(['prep_h'], axis=1)
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(data=court_d_sum, x='charge', y='court_d', hue='Year')
```

```python
# prep_h statistics and prep_h trends on different charges each year
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# prep_h statistics
grouped_prep_h = df.groupby('charge').agg({'prep_h': ['mean', 'min', 'max']})
print(grouped_prep_h)

#prep_h trends on different charges each year
prep_h_sum = df.groupby(['Year', 'charge'], as_index=False).sum()
prep_h_sum = prep_h_sum.drop(['court_d'], axis=1)
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(data=prep_h_sum, x='charge', y='prep_h', hue='Year')
```
